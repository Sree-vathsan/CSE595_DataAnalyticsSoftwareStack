```python
import matplotlib.patches as patches 
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
from shapely.geometry import Polygon

fig1 = plt.figure(figsize=(20, 20)) 
ax1 = fig1.add_subplot(111) 
#fig1, axes1 = plt.subplots(1, 1, figsize=(12,4))
#print(dfSlide1[1:3])
#dfSlide2 = dfSlide1[1:3000]
#boundaries = dfSlide2['Boundaries']
boundaries = dfSlide1['Boundaries']
patchesArr = []
poly_area = []
for bnd in boundaries:
    obj1 = bnd.split(';')
    #print(len(x1))
    xArr = []
    yArr = []
    points = []
    for pts in obj1:
        #print(len(pts))
        if(len(pts) > 0):
            obj2 = pts.split(',')
            points.append((float(obj2[0]),float(obj2[1])))
            xArr.append(obj2[0])
            yArr.append(obj2[1])
    #print(points)
    polygon = Polygon(points)
    poly_area.append(polygon.area)
    #print(points)
    poly = patches.Polygon(points,alpha='0.2',ec='green',hatch='x',linestyle='dashed',fill=False) 
    #ax1.add_patch(poly)
    patchesArr.append(poly)
colors = 100*np.random.rand(len(patchesArr))
p = PatchCollection(patchesArr, cmap=matplotlib.cm.jet, alpha=0.4)
p.set_array(np.array(colors))
ax1.add_collection(p)
ax1.set_xlim(0,700) 
ax1.set_ylim(0,1500)
plt.show()
dfSlide1['Computed_Area'] = poly_area
dfSlide1
```
