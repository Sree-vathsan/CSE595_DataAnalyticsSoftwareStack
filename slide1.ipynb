```python
import matplotlib.patches as patches 
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
from shapely.geometry import Polygon
import numpy as np
import cv2
```

```python
fig1 = plt.figure(figsize=(20, 20)) 
ax1 = fig1.add_subplot(111) 
#fig1, axes1 = plt.subplots(1, 1, figsize=(12,4))
#print(dfSlide1[1:3])
#dfSlide2 = dfSlide1[1:3000]
#boundaries = dfSlide2['Boundaries']
boundaries = dfSlide1['Boundaries']
patchesArr = []
poly_area = []
poly_per = []
poly_bi = []
poly_minEllip_ratio = []
poly_minEllip_rotation = []
poly_minCircleRad = []
poly_nedges = []

for bnd in boundaries:
    obj1 = bnd.split(';')
    #print(len(x1))
    xArr = []
    yArr = []
    points = []
    for pts in obj1:
        #print(len(pts))
        if(len(pts) > 0):
            obj2 = pts.split(',')
            points.append((float(obj2[0]),float(obj2[1])))
            xArr.append(float(obj2[0]))
            yArr.append(float(obj2[1]))
    #print(points)
    polygon = Polygon(points)
    #Feature 1 - Area
    poly_area.append(polygon.area)
    #Feature 2 - Perimeter
    poly_per.append(polygon.length)
    
    #Feature 3 - Border Index
    #Length Calculation, Width Calculation
    #Expression (Perimeter / 2*(length + width))
    lenX = abs(max(xArr) - min(xArr))
    lenY = abs(max(yArr) - min(yArr))
    borderIndex = (polygon.length / 2*(lenX + lenY))
    poly_bi.append(borderIndex)
    
    #Feature 4 - Ratio of length to width of Smallest Enclosing Ellipse 
    minEllip = cv2.fitEllipse(np.asarray(points, np.float32))
    lenWid = minEllip[1]
    poly_minEllip_ratio.append(lenWid[0]/lenWid[1])
    
    #Feature 5 - Rotation of Smallest Enclosing Ellipse 
    poly_minEllip_rotation.append(minEllip[2])
    
    #Feature 6 - Minimum Enclosing Circle's radius
    poly_minCircleRad.append(cv2.minEnclosingCircle(np.asarray(points, np.float32))[1])
    
    #Feature 7 - Compactness
    #The ratio of the area of a polygon to the area of a circle with the same perimeter
    #Calculated at the end
    
    #Feature 8 - No of Edges
    if len(points)==2:
        poly_nedges.append(1)
    else:
        poly_nedges.append(len(points)-1)
    
    #print(points)
    poly = patches.Polygon(points,alpha='0.2',ec='green',hatch='x',linestyle='dashed',fill=False) 
    #ax1.add_patch(poly)
    patchesArr.append(poly)
colors = 100*np.random.rand(len(patchesArr))
p = PatchCollection(patchesArr, cmap=matplotlib.cm.jet, alpha=0.4)
p.set_array(np.array(colors))
ax1.add_collection(p)
ax1.set_xlim(0,700) 
ax1.set_ylim(0,1500)
plt.show()
dfSlide1['Computed_Area'] = poly_area
dfSlide1['Computed Perimeter'] = poly_per
dfSlide1['Border Index'] = poly_bi
dfSlide1['Minimum Ellipse Ratio'] = poly_minEllip_ratio
dfSlide1['Minimum Ellipse Rotation'] = poly_minEllip_rotation
dfSlide1['Minimum Enclosing Circle'] = poly_minCircleRad
dfSlide1['Compactness'] = (np.asarray(poly_area, np.float32) * 4 * np.pi)/(np.asarray(poly_per,np.float32) ** 2)
dfSlide1['Edge Count'] = poly_nedges
#dfSlide1
```
